## Normal (Gaussian) Distribution

1. Bell shaped curve.
2. Lots of things in nature follow Gaussian distribution.

Given a _continuous random variable_ `X`, If `X` has a PDF (Probability Density Function) that looks like a bell shaped curve then `X` has a distribution which is Gaussian distribution.

#### Parameters of Gaussian Distribution 

1. `μ` - Population Mean.
2. `σ²` - Population variance.

Gaussian Distribution is represented as

```
X ~ N(μ, σ²)
```

#### Properties of Gaussian Distribution

1. It has population mean `μ` as zero.  
  ![Normal Distribution](Normal_Distribution.png)
2. As `x` moves away from `μ` `y` reduces `exp(-x^2)`.
3. 68-95-99.7 rule.  
  ![68-95-99-rule](68-95-99-rule.png)

#### Create Distributions using **NumPy**

Syntax:

```python
# Create a normal distribution of population size 500 with mean 5 and standard deviation 4. 
mean = 5
std_dev = 4
pop = 500

np.random.normal(loc=mean, scale=std_dev, size=pop)
```

Create Normal/Gaussian distributions using NumPy `random` module.  

#### Example 1: Using `np.random.normal`.

````python
# X = N(5, 4²)
X = np.random.normal(loc=5, scale=4, size=10)

print("Distribution:")
print(X)

print("\n`loc` Mean:", 5)
print("Actual Mean:", X.mean())
print("Variance:", X.var())

print("Standard Deviation:", X.std())
````

> Note: Distribution will not have exact mean set using `loc`.

#### Variation of bin size

Create normal distribution for continuous random variable.

````python
# X = N(0, 10²)
X = np.random.normal(loc=0, scale=10, size=3000)
````

````python
x_ticks = np.arange(X.min() - (X.min() % 5), X.max() + (X.max() % 5), 5, dtype="int16")
x_ticks
````

Plot the distribution

````python
sns.displot(X, kde=True, bins=10, aspect=2.5)
plt.xlabel("Random variable X")
plt.ylabel("Count")
plt.xticks(ticks=x_ticks)
plt.grid()

sns.displot(X, kind="kde", aspect=2.5)
plt.xlabel("Random variable X")
plt.ylabel("Probability Density Function")
plt.xticks(ticks=x_ticks)
plt.grid()

plt.show()
````

#### CDF: Cumulative Distribution Function

````python
counts, bin_edges = np.histogram(X, bins=10, density=True)

# Compute PDF.
pdf = counts / (sum(counts))
print("PDF:", np.round_(pdf, decimals=2))
print("Bin edges:", np.round_(bin_edges, decimals=2))

# Compute CDF.
cdf = np.cumsum(pdf)

# Set figure width and height.
plt.figure(figsize=(12, 5))

# Plot PDF and CDF of petal_length.
plt.plot(bin_edges[1:], pdf, label="Probability Desnsity Function")
plt.plot(bin_edges[1:], cdf, label="Cumulative Distribution Function")

plt.xlabel("Random Variable X")
plt.ylabel("Probability")
plt.xticks(ticks=x_ticks)
plt.yticks(ticks=np.arange(0, 1.1, 0.1))
plt.grid()
plt.legend()

plt.show()
````
